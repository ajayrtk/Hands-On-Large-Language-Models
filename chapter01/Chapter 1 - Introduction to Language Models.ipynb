# Phi-3

The first step is to load our model onto the GPU for faster inference. Note that we load the model and tokenizer separately (although that isn't always necessary).

**AutoModelForCausalLM**: This is a general class for loading causal language models (used for text generation).

**AutoTokenizer**: This handles tokenization, converting text into numerical format for the model.


**"microsoft/Phi-3-mini-4k-instruct"**: This specifies the Phi-3 Mini model from Microsoft, which is optimized for instruction-following tasks.

**device_map="cuda"**: Loads the model onto a GPU (CUDA) if available, speeding up inference.

**torch_dtype="auto"**: Automatically selects the best data type (e.g., float16 or bfloat16) for efficient performance.

**trust_remote_code=True**: Allows execution of custom model code provided by the model's authors (useful when the model has special configurations).

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

Although we can now use the model and tokenizer directly, it's much easier to wrap it in a `pipeline` object:

The **pipeline** function provides a high-level API for various NLP tasks, including text generation, translation, and summarization.

**text-generation**: Specifies that this pipeline is for text generation.

**model=model**: Uses the previously loaded Phi-3-mini-4k-instruct model.

**tokenizer=tokenizer**: Ensures input text is properly tokenized before being fed into the model.

**return_full_text=False**: Returns only the generated part, excluding the original input.

**max_new_tokens=500**: Limits the number of new tokens the model can generate (prevents excessively long outputs).

**do_sample=False**: Disables randomness, making the output deterministic (uses greedy decoding or beam search instead of sampling).



```python
from transformers import pipeline

# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False
)
```

Finally, we create our prompt as a user and give it to the model:

This is a list of messages structured like a chat history.
{"role": "user", "content": "..."}

*   **role**: "user" indicates that this is input from the user.
*   **content**: "Create a funny joke about chickens." is the actual prompt/question.

The generator function is called with messages as input.

It processes the prompt using the Phi-3 Mini model and generates a response.


```python
# The prompt (user input / query)
messages = [
    {"role": "user", "content": "Create a funny joke about chickens."}
]

# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```
